# DSCI 6001 Final Project Rubric

This short document will provide you some guidance about what you need to turn in to me for your final project. 


### Final Project Overarching Goals

1. Demonstrate that you have an understanding of how to use linear algebra to manipulate data.
2. Work on a project that you find personally interesting and motivating. 
3. Stretch yourself a bit.

You shouldn't be ignoring RATs or your studies in order to knock this out of the park, nor should you be stressing yourself out about it because you're hoping to hang your grade on it. I want to see best efforts while keeping up with material in the rest of the class. 

I have found that projects seem to cluster into a few groups, and so I'll address what each of these need to do:

### People with math backgrounds coding up a single algorithm not taught in class

In general, I expect that you're going to complete the code, that it is bug-free and highly tested. You need to submit a complete explanation of the basic ideas behind the algorithm along with pseudocode for it that other people can follow. Spend a whole 8-12 hour day making a good writeup.

Turn in:
1. a .py file of your completed code, with detailed in-line comments explaining it.
2. a .py file of your unit tests, using either nosetests or unittest (nosetest preferred)
3. A write up of the algorithm in a google doc or docx, its current uses and history, with an explanation of the ideas that motivate the algorithm (figures are a plus) and a nicely written pseudocode.


### People with coding backgrounds coding up several similar algorithms or algorithmic base code and testing them/it


In general, I expect that you're going to complete the code, that it is bug-free and highly tested. Write a synopsis of each of the algorithms or base code you've implemented. For people that just wrote a few algorithms, a two-sentence description of the basic ideas that motivated the algorithm, along with equations if relevant, and pseudocode will suffice. For those that wrote base code submit a synopsis of all the classes you implemented (simple description - one sentence will suffice) and a short discussion of the challenges you overcame and how you overcame them.

Turn in:
1. a .py file of your completed code, with detailed in-line comments explaining it.
2. a .py file of your unit tests, using either nosetests or unittest (nosetest preferred)
3. Brief descriptions of each algorithm or base classes and methods for each in a google doc or docx. In a separate section: Algorithm writers provide pseudocode and relevant equations for each algo. Base code writers discuss design challenges.

### People using a set of already-coded decomposition algorithms to compare or analyze data

In general, your data will be some kind of standard data set. You should have thought about how you're going to compare each of the algorithms and whether or not there is a source of ground truth. What are you comparing them to? Each other? I will be judging based on the kind of thought and the execution you put into this.

Turn in:
1. a .ipynb file of your completed study. You *must* have figures comparing the algorithms and provide a clear graphical representation of your findings in the notebook.
2. A write up n a google doc or docx: 
    a) Discuss the data set you used and why.
    b) A description of what information the algorithms you are using provide
    c) Short descriptions of each algorithm you used, its history and pseudocode for each. Provide equations if relevant.
    d) Results and discussion. Provide figures. Why did you get the result you got? Thoughts for improving the algorithms?

Turn in:
