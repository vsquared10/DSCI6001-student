1) Given the following augmented matrix:
    
$ X = \begin{bmatrix}
3 & 2 & 9\\
1 & 5 & 16
\end{bmatrix}$

Write the associated system of equations and find the solution either by hand or using Numpy.

```python
import numpy as np
alpha = [[3, 2], [1, 5]]
beta = [9, 16]
np.linalg.solve(alpha, beta).reshape(2, 1)
```
