## An introduction to Metrics 

Recall from basic mathematics (and last week's DeMoivre's rules lab) that a "set" is simply a collection of objects (just as it is in python). For example, $X=\{3, 5, 8, 9, 10\}$ is a set, and so is $X=\{'airplane', 'goat' 3x+2y=6\}$. You can have a set of polygons, or a set of turtles, or a set of tools. Anything can be included in a set. We have been working constantly with the set of real numbers, $\mathbb{R}$, which is also a **field** (you have heard me use this concept more), due to the relationship of the numbers in the set with each other (namely that the set is closed under addition and multiplication). Similarly, and inclusively, all $\mathbb{R}^n$ are also fields. 

In data science we are very frequently engaged with the comparison of groups of objects (sets) with each other, for example, given a set of features (feature space), how "similar" are two observations (variable space) to each other?
The notion of comparison is both important and relatively nuanced. For example, suppose we had a data set involving goats (I like goats because I find them to be innately hilarious animals): 

                hoof size   coat length(mm)  weight (kg)  length (cm)  height at shoulder (cm) 
        goat 1     4            20             100         80                   54             
        goat 2     6            30             120         80                   56             
        goat 3     6            30             115         70                   41             


Suppose that we are trying to classify these three goats into two breeds using the above data. 

Using a **metric** we can evoke a relationship among these three points of data, allowing us to determine their relationships. Quite simply, two of the goats must be more similar to each other (of the same breed) than they are to the third.

A metric (meaning "measure" in Latin) is simply a means of measuring "distance" between objects within a given set $X$, providing a single number $n \in \mathbb{R}$:

$X \times X \rightarrow \mathbb{R}$

This number $n$ is the calculated "distance" between the objects inside of $X$. Therefore the metric determines relationships among points in terms of a single value, $n$, in this case. 

Returning to the matter of the goats, two of the goats must be closer to each other (more similar) than they are to a third. But which trait is most important? Coat length? Hoof size? Length? Are all three of these goats from the same or different breeds? Perhaps coat length is not important, as one or more goats have been sheared? Perhaps it is winter or summer. We don't know the sex of the goats. How can we classify them simply, given that we know they must be of one of two breeds?

This leads us to the different types of metrics. One way of creating a metric here might be to simply subtract the values of the features from each other and average the result, i.e.:

            (change in) hoof size   coat length(mm)  weight (kg)  length (cm)  height at shoulder (cm) 
        goat 2-goat 1     2            10             20           0                    2             
        goat 2-goat 3     0            0               5           10                  15             

In order to make these differences into a metric, we might sum them:

        goat 2-goat 1    34             
        goat 2-goat 3    20             

This would suggest that goat 2 and goat 3 are more similar to each other than they are to goat 1. However, this analysis assumes that we can directly compare the units of each feature, and that each feature is equally important. It also assumes that summing negative with positive numbers retains some kind of meaning.

## Definition of a Metric and Types of Metrics

Perhaps the most common and most used concept of a metric is the **Euclidean distance between two vectors $a$ and $b$** in $\mathbb{R}^{n}$:

$$d(a,b) = \sqrt{\sum_{i=1}^{n}(b_i-a_i)^2}$$


This metric expresses the most human-friendly and natural concept of distance between two objects, and sidesteps the inherent complication of *negative-valued distances between two objects.* In fact, this latter issue is a requirement of a metric.

A metric $d(x,y)$ where $(x,y) \in X$  must satisfy four particular conditions to be considered a metric:

1. Non-negativity: $d(x,y) \geq 0$
2. Zero upon equality: $d(x,y) = 0 \iff x = y$
3. Commutativity of arguments: $d(x,y) = d(y,x)$
4. Triangle Inequality: $d(x,z) \leq d(x,y) + d(y,z)$


A metric is said to be *translation invariant* if $d(x,y) = d(x+a,y+a)$ for $x,y,a \in X$

A metric is said to be *rotation invariant* if $d(x,y) = d(T(x),T(y))$ for $x,y \in X$ and $T$ rotation matrix

### Other Common Metrics:

**Manhattan or "Taxicab" Distance, also "Rectilinear distance"**

Measures the relationships between points at right angles, meaning that we sum the absolute value of the difference in vector coordinates. This metric is sensitive to rotation.

$$d_{M}(a,b) = \sum_{i=1}^{n}|(b_i-a_i)|$$

**Chebyschev Distance**

The Chebyschev distance or sometimes the $L^{\infty}$ metric, between two vectors is simply the the greatest of their differences along any coordinate dimension:

$$d_{\infty}(a,b) = \max_{i}{|(b_i-a_i)|}$$


**Cosine Distance**

(note technically Cosine Distance is not a "metric": see if you can figure out which condition in the definition of a metric is violated?)

The cosine distance between two vectors is a measure of rotational relationship rather than magnitude. It is frequently used in bioinformatics where total magnitude changes little as there are often hundreds if not thousands of "coordinates", but relationships among coordinate dimensions may change dramatically.

$$d_{Cos}(a,b) = \frac{a \cdot b}{\|a\|\|b\|}$$


**Hamming or "Rook" Distance**

The hamming distance can be used to compare nearly anything to anything else. It is used in comparison of strings of equal length, and defined as the number of differences in characters between two strings, ie:

$$d_{hamming}('bear', 'beat') = 1$$
$$d_{hamming}('cat', 'cog') = 2$$
$$d_{hamming}('01101010', '01011011') = 3$$

**Edit distance**

This is exactly the same as the hamming distance, but also includes insertions and deletions, and so can compare strings of any length to each other. 

$$d_{edit}('lead', 'gold') = 4$$
$$d_{edit}('monkey', 'monk') = 2$$
$$d_{edit}('lucas', 'mallori') = 8$$

## Norms and Lebesgue space


The $L^{p}$, or Lebesgue space, consists of all of functions of the form:

$\|f\|_{p} = (|x_{1}|^{p}+|x_{2}|^{p}+|x_{n}|^{p}+)^{1/p}$

This mathematical object is known as a *p-norm*. You are already at least cursorily familiar with p-norms, as the most common measure of distance of a vector with respect to itself (it's length) is a norm.


### $p$-norms

For any $n$-dimensional real or complex vector. i.e. $x \in \mathbb{R}^n \text{ or } \mathbb{C}^n$

$$ \|x\|_p = \left(|x_1|^p+|x_2|^p+\dotsb+|x_n|^p\right)^{\frac{1}{p}} $$

$$ \|x\|_p = \begin{pmatrix}\sum_{i=1}^n{|x_i|^p} \end{pmatrix}^{\frac{1}{p}} $$


### Norm properties

Much like metrics, all $p$-norms satisfy a set of four properties:

For a set $V$ defined over a field $F$ a _norm_, denoted as $\|.\|$, is a function from $V$ to $R$ that satisfies the following conditions for any ${x, y} \in V$ and $\text{a} \in R$:

1. Non-negativity $ \|x\| > 0 \ \forall x \neq 0 $
2. Mapping of additive identity $ \|0\| = 0 $
3. Absolute scalability $ \|ax\| = |a|\|x\| $
4. Triangle inequality $ \|x + y \| \leq \|x\| + \|y\| $

If the set $V$ represents a vector space, then the vector space taken together with a norm, $(V,\|.\|)$, is called a **normed vector space**. Given a normed vector space $(V,\|.\|)$ and vectors ${x, y} \in V$, and we define a metric: 

$$ d(x,y) = \|x-y\| $$

For vectors the norm specifies the *length* of the vector; the value of *length* given by the norm varies based on the type of norm chosen. 

The the Euclidean length of a vector (directly related to the Euclidean distance) is the Euclidean norm. This is the same thing as the $L^{2}$ norm:

$\|f\|_{2} = (|x_{1}|^{2}+|x_{2}|^{2}+|x_{n}|^{2}+)^{1/2}$

## The Unit Circle

 The **Unit circle**, or *norm ball*, of a norm is a unique characteristic of a norm that helps us visualize the norm and understand the shape, or *topology*, of the underlying normed space. The unit circle of an L$^p$-norm in $\mathbb{R}^n$ is given by: 

$$\{ x \in \mathbb{R}^n \ |\  \|x\|_p = 1 \}$$

What this means is that the points allowable within the unit ball are those whose norms are less than or equal to 1 (hence the "unit" moniker). The insight provided by this concept is that an appropriate choice of translation and rotation can transform any sphere into a unit sphere. 

The type of norm being used, however, changes the relative shape of the ball. Thus, the unit ball is a simple summary of the relationships captured by the chosen norm. Just like the choice of a distance metric (e.g. Manhattan vs. Euclidean distance), the choice of a norm profoundly changes the way we represent information.   

We can plot the unit circle of the Euclidean norm (in 2-space) as follows:

```python
import pylab
from numpy import array, linalg, random, sqrt, inf
import matplotlib.pyplot as plt
%matplotlib inline

def plotUnitCircle(p, n):
    
    #given a number of requested points
    for i in range(n):
        
        #select a random point x, with result centered on 0,0
        x = array([random.rand()*2-1,random.rand()*2-1])
        
        #this restricts the sample to fit the norm
        if linalg.norm(x,p) < 1:
            plt.plot(x[0],x[1],'bo')
    plt.show()
    
plotUnitCircle(2,5000)
```

### Unit Circle with increasing p

The shape of the unit circle (ball) describes the distance relationships permitted by the norm. If $p$ is very small, the importance of the smallest dimensions of $\vec{x}$ are increasingly small, and the value of the norm is primarily determined by the largest variable. As $p$ increases, the norm is increasingly dependent on both dimensions. Increasing $p$ begins to give more weight (relative to the rest of the entries $x_i$) to the smaller values of $\vec{x}$. 
At $$\lim_{p \rightarrow \infty} L^p$$

the norm becomes $\|\vec{x}\|_{\infty} = max\{|x_1|,|x_2\, ... |x_n| \}$, in other words the largest $|x_i|\leq 1$, describing a square. For two vectors $a$ and $b$, this would mean that we simply take the greatest difference within a given feature between the two of them. Note the direct relationship between this norm and the Chebyshev metric. This is because norms and metrics are directly related, the only difference being that the metric is a norm of a difference between two objects, whereas the norm simply measures the length of an object itself.
